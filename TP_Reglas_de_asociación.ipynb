Accedemos al dataset

```
import pandas as pd

# indicamos cual dataset queremos utilizar, en este caso el nro. 24
dataset = openml.datasets.get_dataset(24)

# separamos las información almacenada en el dataset
X, y, categorical_indicator, attribute_names = dataset.get_data(
    dataset_format='dataframe',
    target=dataset.default_target_attribute
)

#  concatenamos la información relevante en un único DataFrame 
df = pd.concat([X, y], axis=1)
df
```

```
df3 = (df.iloc[:,[4,5,6,7,8,18,19,20,21,22]])
df3
```

```
df_dummies = pd.get_dummies(df3)  #le indicamos las columnas para transformar
df_dummies
```

```
from mlxtend.frequent_patterns import apriori    #funcion apriori se importa del mlxtend

apriori(df_dummies, min_support=0.3) 
```

```
apriori(df_dummies, min_support=0.3, use_colnames=True)
```

```
frequent_itemsets = apriori(df_dummies, min_support=0.3, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x)) #le agrego una columna de la longitud del itemset frecuente para despues filtrar o instancias
frequent_itemsets
```

```
frequent_itemsets[ (frequent_itemsets['length'] >= 2) &      #elijo con que itemset me quedo
                   (frequent_itemsets['support'] >= 0.3) ]
```

```
from mlxtend.frequent_patterns import association_rules                                                                                                           #association rules genera las reglas a partir de la estructura y mira el soporte asociado, necesita el soporte del antecedente

# utilizamos confianza = 0.4
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules
```

```
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x)) #agrego el tamaño del antecedente
rules
```

```
rules[ (rules['antecedent_len'] >= 2) &    #estrategia de filtrado que tenga 2 o mas antecedente y el confidente mayor a 0.75
       (rules['confidence'] > 0.80) &
       (rules['lift'] > 1) ]
```
