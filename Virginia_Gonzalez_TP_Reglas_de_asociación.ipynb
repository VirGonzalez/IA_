Accedemos al dataset

```
import pandas as pd

# indicamos cual dataset queremos utilizar, en este caso el nro. 24
dataset = openml.datasets.get_dataset(24)

# separamos las informaciÃ³n almacenada en el dataset
X, y, categorical_indicator, attribute_names = dataset.get_data(
    dataset_format='dataframe',
    target=dataset.default_target_attribute
)

#  concatenamos la informaciÃ³n relevante en un Ãºnico DataFrame 
df = pd.concat([X, y], axis=1)
df
```

```
df.columns
```

```
df = df[['population', 'habitat']]
df
```

```
df_dummies = pd.get_dummies(df, prefix=['pop', 'hab'],columns=['population', 'habitat'])
df_dummies
```

**Objetivo**


El objetivo de este trabajo sera determinar si existe una relacion entre el habitat y la disposicion de la poblacion debido a la interaccion con compuestos quimicos del ambiente. El microbioma de la atmosfera puede afectar al hongo y a traves de este trabajo buscaremos dilucidar si existe un efecto del habitat.
Espero encontrar que en aquellos habitats mas disturbados como urbanos la disposicion de los hongos sera mas agregada, mientras en bosque y pastizales expandiran su area de ocupacion



```
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(df_dummies).transform(df_dummies)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```
from mlxtend.frequent_patterns import apriori

apriori(df , min_support=0.1,use_colnames=True)
```
